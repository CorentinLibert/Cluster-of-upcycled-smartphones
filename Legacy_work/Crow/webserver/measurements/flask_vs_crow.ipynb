# Performances: Flask vs Crow
This notebook show performance of a simple web application that receive an image, download, copy it to another image (by reading it) and then send the copy back. The app has been developed in Flask (python) and in Crow (c++).

Not that the version done in Crow may be less performant compared to the one in flask due to the fact that more things have be done by hand.

## Results

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = "flask_vs_crow.csv"
df = pd.read_csv(data)

# display(df)

plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")
sns.boxplot(data=df, x="process", y="time_ns", hue="framework", palette="Dark2")
plt.title("Comparison of the execution time (in nanoseconds) for each process from the web app for Flask and Crow")
plt.xlabel("Processes")
plt.ylabel("Time (in nanoseconds)")
plt.legend(title="Framework")
# plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
plt.tight_layout()
plt.savefig("comparison_flask_vs_crow_webserver.jpg")
plt.savefig("comparison_flask_vs_crow_webserver.png")
plt.show()
```

```python
mask = df["process"] == "processing"
df_without_processing = df[~mask]

plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")
sns.boxplot(data=df_without_processing, x="process", y="time_ns", hue="framework", palette="Dark2")
plt.title("Comparison of the execution time (in nanoseconds) for each process (except 'processing') from the web app for Flask and Crow")
plt.xlabel("Processes")
plt.ylabel("Time (in nanoseconds)")
plt.legend(title="Framework")
# plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
plt.tight_layout()
plt.savefig("comparison_flask_vs_crow_webserver_without_processing.jpg")
plt.savefig("comparison_flask_vs_crow_webserver_without_processing.png")
plt.show()
```
